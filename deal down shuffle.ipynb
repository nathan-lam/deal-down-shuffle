The deal down shuffle is a methematical way to shuffle a packet of card.
The steps to perform this is:
1. place top card down in a separate pile
2. place next card on the bottom of the dealing packet
3. place the next card on top of the separate pile
4. repeat until no cards are left in the hand.

```
import numpy as np

def T(x):
  '''
  [1,2,3,4] -> [4,2,3,1]
  [1,2,3,4,5] -> [2,4,5,3,1]
  [1,2,3,4,5,6] -> [4,6,2,5,3,1]
  '''
  if x == []:
    return "bad array"
  elif len(x) == 1:
    return x
  else:
    y = [x.pop(0)]
    x.append(x.pop(0))
    return T(x) + y
```

```
from math import log, floor, ceil
def alpha(N,x):
  # N = number of cards
  # x = old position
  # return power constant for calculation
  # this is an approximation
  A = log(N,2) - log(2*x-1,2) - 0.5
  if abs(floor(A)-A) <= 0.5:
    return floor(A)
  else:
    return ceil(A)

 # return (log(N,2) - log(2*x-1,2) - 0.5)
```

```
def N(n,x):
  # n = number of cards
  # x = old position
  # return new position
  a = alpha(n,x)
  print((2*x-1)*pow(2,a))
  print(a)
  if n > (2*x-1)*pow(2,a):
    return int(2*(n-(2*x-1)*pow(2,a)))
  elif n <= (2*x-1)*pow(2,a):
    return int(2*n-(2*x-1))
  else:
    print("Error")
```

```
def num_to_reset(N, max_iter = 100):
  #returns the number of deals needed to reset
  start = list(range(1,N+1))
  looping = list(range(1,N+1))
  iter = 0
  for i in range(max_iter):
    looping = T(looping)
    iter += 1
    if looping == start:
      return iter

  return iter
```
